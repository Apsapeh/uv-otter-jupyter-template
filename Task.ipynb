## Задание 1 (2 балла): Сумма двух чисел

Напишите функцию `add(a, b)`, которая возвращает сумму двух чисел `a` и `b`.

```python
def add(a, b):
    return a + b  # SOLUTION
```

```python
# Открытый тест (студент видит)
assert add(2, 3) == 5
assert add(0, 0) == 0
```

```python
# HIDDEN
assert add(-1, 1) == 0
assert add(-5, -3) == -8
assert add(100, 200) == 300
```

## Задание 2 (3 балла): Факториал

Напишите функцию `factorial(n)`, которая возвращает факториал числа `n`.

Считайте, что `n >= 0`. Напоминание: `0! = 1`.

```python
def factorial(n):
    if n == 0:  # SOLUTION
        return 1  # SOLUTION
    return n * factorial(n - 1)  # SOLUTION
```

```python
# Открытый тест (студент видит)
assert factorial(5) == 120
```

```python
# HIDDEN
assert factorial(0) == 1
assert factorial(1) == 1
assert factorial(10) == 3628800
```
